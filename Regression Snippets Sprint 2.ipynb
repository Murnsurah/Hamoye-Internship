From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the Root Mean Squared error in three D.P?

```python
#preparing the training data
x = df[['T2']]
y= df['T6']
```

```python
#split your data into test and training data
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = 42)
```

```python
#apply model to Linear Regression
model = LinearRegression()
```

```python
#fitting your model
model.fit(x_train,y_train)
```

```python
#evaluation of model
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Slope : {round(model.coef_[0],2)}")
print(f"Intercept = {round(model.intercept_,2)}")
print(f"MSE = {round(mse,2)}")
print(f"r2_score = {r2}")
```

```python
rmse = np.sqrt(mse)
```

Remove the following columns: [“date”, “lights”]. The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42 (for reproducibility). Normalize the dataset using the MinMaxScaler (Hint: Use the MinMaxScaler fit_transform and transform methods on the train and test set respectively). Run a multiple linear regression using the training set

```python
x = df.drop(columns=['date','lights','Appliances'])
y= df['Appliances']
```

```python
x
```

```python
# Split the dataset into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model = LinearRegression()
model.fit(x_train_scaled, y_train)
```

```python
# making predictions
y_pred_train = model.predict(x_train_scaled)
y_pred_test = model.predict(x_test_scaled)
```

```python
# calculating RMSE
rmse_train = mean_squared_error(y_train, y_pred_train, squared=False)
rmse_test = mean_squared_error(y_test, y_pred_test, squared=False)
mae_train = mean_absolute_error(y_train, y_pred_train)


print("Mean Absolute Error on training set:", round(mae_train, 3))
print("RMSE on training set:", round(rmse_train,3))
print("RMSE on test set:", round(rmse_test,3))
```

What is the Mean Absolute Error (in three decimal places) for the  training set?

```python
print("Mean Absolute Error on training set:", round(mae_train, 3))
```

What is the Root Mean Squared Error (in three decimal places) for the training set?

```python
print("RMSE on training set:", round(rmse_train,3))
```

What is the Mean Absolute Error (in three decimal places) for test set?

```python
print("RMSE on test set:", round(rmse_test,3))
```

What is the Root Mean Squared Error (in three decimal places) for test set?

```python
print("RMSE on test set:", round(rmse_test,3))
```

Train a ridge regression model with default parameters. Is there any change to the root mean squared error (RMSE) when evaluated on the test set?

```python
from sklearn.linear_model import Ridge

# Train the Ridge regression model
ridge_model = Ridge()
ridge_model.fit(x_train_scaled, y_train)

# Make predictions on the test set
y_pred_test_ridge = ridge_model.predict(x_test_scaled)

# Calculate RMSE for the Ridge regression model on the test set
rmse_test_ridge = mean_squared_error(y_test, y_pred_test_ridge, squared=False)

print("RMSE on test set (Ridge regression):", round(rmse_test_ridge, 3))
print('The RMSE for the Ridge regression model is only slightly higher than the RMSE,This suggests Ridge regression did not significantly improve the model')
```

Train a lasso regression model with default value and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
# Training the Lasso regression model
lasso_model = Lasso()
lasso_model.fit(x_train_scaled, y_train)

# Get the feature weights
feature_weights = lasso_model.coef_

# Print the feature weights
print("Feature weights:", feature_weights)
num_nonzero_features = sum(feature_weights != 0)

print("Number of features:", num_nonzero_features)
```

What is the new RMSE with the Lasso Regression on the test set?

```python
# Make predictions on the test set using the Lasso regression model
y_pred_test_lasso = lasso_model.predict(x_test_scaled)

# Calculate RMSE for the Lasso regression model on the test set
rmse_test_lasso = mean_squared_error(y_test, y_pred_test_lasso, squared=False)

print("RMSE on test set (Lasso regression):", round(rmse_test_lasso, 3))
```
